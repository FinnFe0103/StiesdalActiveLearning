```python
first_poly = results_svr[results_svr['Combination'].str.contains('poly')].iloc[0]
first_rbf = results_svr[results_svr['Combination'].str.contains('rbf')].iloc[0]
first_sigmoid = results_svr[results_svr['Combination'].str.contains('sigmoid')].iloc[0]
print(first_poly['Combination'])
print(first_rbf['Combination']) 
print(first_sigmoid['Combination'])
```

```python
best_svr_kernels = [first_poly['Combination'], first_rbf['Combination'], first_sigmoid['Combination']]

kernel_svr = results_svr[results_svr['Combination'].isin(best_svr_kernels)].sort_values(by='Exploration + Exploitation Steps required', ascending=True)

legend_name_mapping = {
    first_poly['Combination']: 'Poly',
    first_rbf['Combination']: 'RBF',
    first_sigmoid['Combination']: 'Sigmoid',
}

kernel_svr['Combination'] = kernel_svr['Combination'].map(legend_name_mapping)

kernel_svr.head(10)
```

```python
# Creating a figure with four subplots (2x2)
fig, axes = plt.subplots(1, 2, figsize=(20, 6), sharex='col')
plt.rcParams["font.family"] = "Times New Roman"

# First row, SVR data
sns.lineplot(ax=axes[0], data=kernel_svr, x='Step_Value', y='Exploitation Stepts required', hue='Combination', marker='o')
axes[0].xaxis.set_major_locator(plt.MaxNLocator(integer=True))
axes[0].set_title('SVR Kernels: Steps vs. Exploitation Steps required', fontsize=14)
axes[0].set_xlabel('Steps', fontsize=14)
axes[0].set_ylabel('Exploitation Steps required', fontsize=14)
axes[0].legend(title='Kernel', fontsize=11)

sns.lineplot(ax=axes[1], data=kernel_svr, x='Step_Value', y='Exploration + Exploitation Steps required', hue='Combination', marker='o')
axes[1].xaxis.set_major_locator(plt.MaxNLocator(integer=True))
axes[1].set_title('SVR Kernels: Steps vs. Total Steps', fontsize=14)
axes[1].set_xlabel('Steps', fontsize=14)
axes[1].set_ylabel('Exploration + Exploitation Steps required', fontsize=14)
axes[1].legend(title='Kernel', fontsize=11, loc='upper right')

axes[1].axvline(x=5, color='#1f77b4', linestyle='--')
axes[1].text(5.2, 225, 'Low Point \n Poly (12)', horizontalalignment='left', color='#1f77b4', fontsize=11) #blue #1f77b4

axes[1].axvline(x=8, color='#ff7f0e', linestyle='--')
axes[1].text(8.2, 200, 'Low Point \n RBF (19)', horizontalalignment='left', color='#ff7f0e', fontsize=11) #orange

axes[1].axvline(x=20, color='#47A02E', linestyle='--')
axes[1].text(19.8, 180, 'Low Point \n Sigmoid (405)', horizontalalignment='right', color='#47A02E', fontsize=11) #green #47A02E
```
